```python
import numpy as np
import matplotlib.pyplot as plt

# Define Softmax function
def softmax(x):
    # Exponentiate the input and normalize by the sum of exponentials
    exp_values = np.exp(x - np.max(x))  # Subtract max(x) for numerical stability
    return exp_values / np.sum(exp_values)

# Generate input values (logits) for 3 classes
x = np.linspace(-2, 2, 10) 
logits = np.array([x, x * 0.5, x * -0.5])  

# Apply Softmax to each row (each class)
softmax_outputs = np.apply_along_axis(softmax, 1, logits)

# Plot the Softmax outputs for each class
plt.figure(figsize=(8, 4))
for i in range(softmax_outputs.shape[0]):
    plt.plot(x, softmax_outputs[i], label=f'Class {i+1}')


plt.title('Softmax Activation Function')
plt.xlabel('Logits (input values)')
plt.ylabel('Softmax Output')
plt.legend()
plt.grid(True)
plt.show()
```
