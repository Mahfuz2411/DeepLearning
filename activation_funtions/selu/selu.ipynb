# SELU (Scaled Exponential Linear Unit) Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define SELU function
def selu(x, alpha=1.67326, lambda_=1.0507):
    return lambda_ * np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Generate input values
x = np.linspace(-10, 10, 100)
y = selu(x, alpha=1.67326, lambda_=1.0507)

# Plot the SELU function
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='SELU (α=1.67326, λ=1.0507)', color='red')
plt.title('SELU Activation Function')
plt.xlabel('x')
plt.ylabel('SELU(x)')
plt.grid(True)
plt.legend()
plt.show()
```
